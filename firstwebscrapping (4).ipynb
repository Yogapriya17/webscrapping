```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.gadgetsnow.com/laptops/HCL"
```

```python
url
```

```python
allow=requests.get(url)
```

```python
allow
```

```python
soup=BeautifulSoup(allow.text)
```

```python
soup
```

```python
item_name=soup.select("._1M-Zt")
```

```python
item_name
```

```python
object_name=[]
```

```python
for i in item_name:
    object_name.append(i.text)
```

```python
object_name
```

```python
price=soup.select("._3tTel")
```

```python
price
```

```python
computer_price=[]
```

```python
for j in price:
    computer_price.append(j.text)
```

```python
computer_price
```

```python
import pandas as pd
```

```python
gadgets_data=pd.DataFrame(object_name,columns=["product_name"])
```

```python
gadgets_data
```

```python
gadgets_data["price_product"]=computer_price
```

```python
gadgets_data
```

```python
storage=soup.select(".laptop_storage+ .S0qxL")
```

```python
storage
```

```python
comp_storage=[]
```

```python
for k in storage:
    comp_storage.append(k.text)
```

```python
comp_storage
```

```python
gadgets_data["lap_storage"]=comp_storage
```

```python
gadgets_data
```

```python
gadgets_data.to_csv("My_first_web_scrapper.csv")
```
